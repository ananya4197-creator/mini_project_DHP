## 1. Calculate total revenue by plan type.

```python
df.groupby("Contract")["MonthlyCharges"].sum()
```

## 3. Calculate churn rate.


```python
churn_rate = (df["Churn"] == "Yes").mean() * 100
print("Churn Rate:", churn_rate)
```
